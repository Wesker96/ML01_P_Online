### Задача:
Научиться работать с видео с помощью OpenCV.
- Записать видео и преобразовать его так что бы оно поворачивалось на 10 градусов в секунду
- Преобразованное видео вернуть к исходному
- Сконвертировал видео в формат mp4 при помощи ffmpeg

### Решение:
- Написал функцию которая вращает видео от центра на определенный градус.
- Написал функцию вращения видео на ввод которой передается видео, которое нужно преобразовать, градусы вращения в секунду и название для записи.
- Если передать 0 вместо название видео будет записано с помощи VideoCapture
- Сконвертировал видео в H.265(.mp4) на сайте https://www.convertfiles.com/

### Вывод:
